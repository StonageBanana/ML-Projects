The trend is then displayed in the chart's title.

```python
import os
import plotly.graph_objs as go
import plotly.io as pio
```

```python
def cv_graphs(company):
  CSV_URL = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol=' + company + '&interval=5min&slice=year1month1&apikey=UQIAWPA3R7MT2NP7'

  with requests.Session() as s:
      download = s.get(CSV_URL)
      decoded_content = download.content.decode('utf-8')
      cr = csv.reader(decoded_content.splitlines(), delimiter=',')
      my_list = list(cr)

  header = my_list[0]
  data = my_list[1:]
  df = pd.DataFrame(data, columns=header)

  df['time'] = pd.to_datetime(df['time'])
  df = df.sort_values(by='time')
  df['open'] = pd.to_numeric(df['open'])
  df['high'] = pd.to_numeric(df['high'])
  df['low'] = pd.to_numeric(df['low'])
  df['close'] = pd.to_numeric(df['close'])

  working_hours = (df['time'].dt.hour >= 9) & (df['time'].dt.hour <= 16)
  df = df[working_hours]
  dates = df['time'].dt.date.unique()

  # Create the directory to store the graphs
  if not os.path.exists(company):
      os.mkdir(company)

  count = 0
  for date in dates:
      sub_df = df[df['time'].dt.date == date]
      candlestick = go.Candlestick(x=sub_df['time'], open=sub_df['open'], high=sub_df['high'], low=sub_df['low'], close=sub_df['close'])
      ma = sub_df['close'].rolling(window=10).mean()
      trend_line = go.Scatter(x=sub_df['time'], y=ma, line=dict(color='red'), name='Trend Line')
      fig = go.Figure(data=[candlestick, trend_line])
      fig.update_layout(title=f'Candlestick chart with Trend Line for {date}')
      fig.show()
      file_path = f"{company}/{count}.html"
      pio.write_html(fig, file_path, auto_open=False)
      count = count + 1
```
