The Maslach Burnout Inventory (MBI) is a psychological assessment tool designed to measure burnout, which is a state of emotional, mental, and physical exhaustion caused by prolonged and excessive stress.

The MBI consists of 22-25 items, depending on the version used, and is typically completed by individuals who are experiencing burnout or are at risk of burnout, such as healthcare professionals, social workers, and teachers. The inventory measures three components of burnout: emotional exhaustion, depersonalization, and personal accomplishment.

* **MBI Emotional Exhaustion** is measured through items such as "I feel emotionally drained from my work" and "I feel frustrated by my job". Higher scores on the emotional exhaustion subscale of the MBI indicate higher levels of emotional exhaustion and are associated with negative outcomes such as decreased job satisfaction, decreased job performance, and increased risk of physical and mental health problems.

* **MBI Cynicism** is measured through items such as "I've become more callous towards people since starting this job" and "I feel I'm treating some recipients as if they were objects". Higher scores on the cynicism subscale of the MBI indicate higher levels of cynicism and are associated with negative outcomes such as decreased job satisfaction, decreased job performance, and increased risk of physical and mental health problems.

*  **MBI Academic Efficacy** is measured through items such as "I feel I'm making an impact through my work" and "I feel I'm successful in meeting the goals of my job". Higher scores on the personal accomplishment subscale of the MBI indicate higher levels of personal accomplishment and are associated with positive outcomes such as increased job satisfaction and better mental and physical health.

The **Center for Epidemiologic Studies Depression Scal**e (CES-D) is a self-report questionnaire designed to measure depressive symptoms in the general population. The scale was developed by the National Institute of Mental Health (NIMH) and has been widely used in research and clinical settings.

The CES-D consists of 20 items, each rated on a 4-point scale (0-3), with higher scores indicating greater depressive symptomatology. The items assess various domains of depression, including sadness, lack of appetite, sleep disturbance, and difficulty concentrating. The scale takes approximately 10-20 minutes to complete.

The **State-Trait Anxiety Inventory** (STAI) is a self-report questionnaire that assesses two types of anxiety: state anxiety and trait anxiety. State anxiety refers to feelings of anxiety that are temporary and situation-specific, whereas trait anxiety refers to a stable tendency to experience anxiety across situations.

The STAI consists of two separate scales: the State Anxiety Scale (SAS) and the Trait Anxiety Scale (TAS). Each scale contains 20 items, each rated on a 4-point scale (1-4), with higher scores indicating greater anxiety. The SAS assesses how the individual feels "right now, at this moment," while the TAS assesses how the individual generally feels.

### Finding relationships between Language spoken, Burnout Levels using Chi-Square Test of Independence


Finding relationships between 2 categorical variables : Chi-Square Test of Independence

* Null hypothesis: There is no relationship between the two variables
* Alternative hypothesis: There is a relationship between the two variables
If the p-value is less than 0.05, we reject the null hypothesis and conclude that there is a relationship between the two variables

```python
# Chi-Square Test of Independence for each pair of categorical variables in a new dataframe
from scipy.stats import chi2_contingency
chi2_table = pd.DataFrame(columns=['Variable 1', 'Variable 2', 'Chi-Square', 'p-value'])
for i, col1 in enumerate(d.columns):
    for j, col2 in enumerate(d.columns):
        if i < j:
            chi2, p, dof, ex = chi2_contingency(pd.crosstab(d[col1], d[col2]))
            chi2_table = chi2_table.append({'Variable 1': col1, 'Variable 2': col2, 'Chi-Square': chi2, 'p-value': p}, ignore_index=True)
```

```python
# Keep only the pairs of variables that are related (p-value < 0.05)
#and sort them by p-value in ascending order (the smaller the p-value, the stronger the relationship)
chi_table = chi2_table[chi2_table['p-value'] < 0.05].sort_values(by='p-value')
chi_table
```

### Principal Component Analysis (PCA)

PCA is a dimensionality reduction technique that is used to reduce the number of variables in a dataset while retaining as much information as possible. It is used to reduce the number of variables in a dataset by transforming a large set of variables into a smaller one that still contains most of the information in the large set.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the data
scaler = StandardScaler()
data_cat_scaled = scaler.fit_transform(data_cat)

# Create a PCA instance: pca
pca = PCA(n_components=0.95)

# Fit the PCA instance to the scaled samples
pca.fit(data_cat_scaled)

# Transform the scaled samples: pca_features
pca_features = pca.transform(data_cat_scaled)

# Print the shape of pca_features
print("The shape of the pca is : ", pca_features.shape)
```

```python
# Plot the cumulative sum of the explained variance ratio
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

Interpretation:
1. The explained variance tells you how much information (variance) can be attributed to each of the principal components.
2. The cumulative explained variance tells you how much information (variance) can be attributed to the principal components.

#### Loadings of the principal components

Loadings are the coefficients that describe how each variable contributes to the principal components. The loadings are the weights that are assigned to each variable in a principal component. The higher the absolute value of the loading, the more important the variable is in that principal component. The loadings can be positive or negative, depending on the direction of the correlation between the variable and the principal component.

```python
# Create a dataframe of the loadings of the first four principal components and the variable names (columns) of the original dataset (index)
loadings = pd.DataFrame({'Feature': data_cat.columns, 'PC1': pca.components_[0], 'PC2': pca.components_[1], 'PC3': pca.components_[2], 'PC4': pca.components_[3]})
loadings

# Sort the loadings of each principal component by their absolute value in descending order and print the first 5 rows
loadings = loadings.sort_values(by=['PC1', 'PC2', 'PC3', 'PC4'], ascending=False)
loadings.head(20)
```

The most significant original variables for each PC can be identified by looking at the magnitude of their loadings, or the weights assigned to each variable in the construction of each PC.

* For the first PC, the variables with the highest absolute loadings are "cesd", "stai_t", and "mbi_ex".
* For the second PC, the variable with the highest absolute loading is "qcae_cog".
* For the third PC, the variable with the highest absolute loading is "amsp".
* For the fourth PC, the variable with the highest absolute loading is "age".

These are the most significant original variables for each PC, and they are the ones that contribute the most to explaining the variance in the data set.
